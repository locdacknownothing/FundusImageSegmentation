```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_nilabel(filename: str, slice_index: int, is_label: bool, title: str = ""):
    # 1. Load the file
    nifti_img = nib.load(filename)

    # 2. Get the data as a numpy array
    # The shape is typically (Height, Width, Depth)
    data = nifti_img.get_fdata()
    print(f"Shape of volume: {data.shape}")

    if is_label:
        print(f"Unique labels found (classes): {np.unique(data).astype(int).tolist()}")

    # 3. Visualize one slice
    slice_data = data[:, :, slice_index]
    plt.figure(figsize=(6, 6))
    # Use 'nearest' interpolation to keep edges sharp for segmentation masks
    plt.imshow(slice_data, cmap='jet', interpolation='nearest')
    plt.colorbar(label='Class Label')
    plt.title(title)
    plt.axis('off')
    plt.show()
```
